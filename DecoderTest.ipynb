```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Caminho onde seu modelo fine-tuned foi salvo
model_path = "/home/vbertalan/Downloads/Modelos/gpt2_log_model"

# Carregando tokenizer e modelo
tokenizer = GPT2Tokenizer.from_pretrained(model_path)
model = GPT2LMHeadModel.from_pretrained(model_path)

# Coloque o modelo em modo de avaliação
model.eval()
model.to("cuda" if torch.cuda.is_available() else "cpu")
```

```python
prompt = "make error"
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

#Generate the next tokens (max of 30 new tokens, for example)

outputs = model.generate(
    **inputs,
    max_new_tokens=50,                      # number of new tokens to generate (range: 5–50; 10–30 for short logs)
    do_sample=True,                         # True = sampling (more creative), False = deterministic
    top_k=50,                               # consider only top K tokens at each step (range: 0 or 10–100; 20–50 for logs)
    top_p=0.95,                             # nucleus sampling cumulative probability (range: 0.8–1.0; 0.9 is balanced)
    temperature=0.8,                        # randomness control (range: 0.1–2.0; 0.7–1.0 for stable logs)
    eos_token_id=tokenizer.encode("\n")[0], # stops when newline appears
    pad_token_id=tokenizer.eos_token_id
)


# Decodificar e imprimir
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(generated_text)
```
